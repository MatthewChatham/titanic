```python
# Imports and setup
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%xmode plain
```

## Descriptives

```python
train.head()
```

Here we see 11 features and the target variable. These features are as follows:

* `PassengerID`: ID used for Kaggle competition
* `Survived`: 0 for deceased, 1 for survivor
* `Pclass`: The passenger class, where 1=First, 2=Second, and 3=Third (proxy for SES)
* `Name`: The passenger's name, of the form (last), (title) (first) (middle) ((wife's name))
* `Sex`: The passenger's sex, male and female
* `Age`: Passenger age in years
* `SibSp`: The number, combined, of siblings and spouses on board the ship
* `Parch`: The number, combined, of (blood) parents and children on board the ship
* `Ticket`: Ticket number
* `Fare`: Passenger fare
* `Cabin`: Passenger's cabin number (First Class only)
* `Embarked`: Point of embarkation, where S=Southampton, C=Cherboug, and Q=Queenstown

Let's transform the categorical variables:

```python
# Transform class to categories
train['Pclass'] = train['Pclass'].replace({1: 'Upper', 2: 'Middle', 3: 'Lower'})

# Training data
train['Pclass'] = pd.Categorical(train['Pclass'], 
                                 categories=['Lower', 'Middle', 'Upper'], 
                                 ordered=True)

train['Sex'] = pd.Categorical(train['Sex'], 
                                 categories=['male', 'female'], 
                                 ordered=False)

train['Embarked'] = pd.Categorical(train['Embarked'], 
                                 categories=['S', 'C', 'Q'], 
                                 ordered=True) # Order based on route

test['Pclass'] = pd.Categorical(test['Pclass'], 
                                 categories=['Lower', 'Middle', 'Upper'], 
                                 ordered=True)

# Test data
test['Sex'] = pd.Categorical(test['Sex'], 
                                 categories=['male', 'female'], 
                                 ordered=False)

test['Embarked'] = pd.Categorical(test['Embarked'], 
                                 categories=['S', 'C', 'Q'], 
                                 ordered=True) # Order based on route

# Cut `Age` and `Fare`
cut_points = [0, 10, 20, 30, 40, 50, 60, 70, 80]
lbls = ['{}-{}'.format(cut_points[i], cut_points[i+1]) 
        for i, val in enumerate(cut_points) if i < len(cut_points) - 1]
binned_age = pd.cut(train['Age'],
       bins=cut_points,
       labels=lbls)
train['age_decade'] = pd.Categorical(binned_age,
                                     categories=lbls, ordered=True)

cut_points = [i*50 for i in range(12)]
lbls = ['{}-{}'.format(cut_points[i], cut_points[i+1])
        for i, val in enumerate(cut_points) if i < len(cut_points) - 1]
binned_fare = pd.cut(train['Fare'],
       bins=cut_points,
       labels=lbls)
train['fare_binned'] = pd.Categorical(binned_fare,
                                     categories=lbls, ordered=True)
```

```python
def prop_missing(series):
    return series.isnull().sum() / series.isnull().size

train.groupby('Embarked').aggregate(prop_missing)[['Age', 'Cabin']]
```

Here we see that almost two-thirds of those who embarked at Queenstown have no age recorded, compared to less than a fifth for those from Southampton. This, along with the fact that 95% of passengers from Queenstown have no recorded cabin number, leads me to believe that these are largely lower-class passengers.

Indeed, among those who embarked at Queenstown, 94% are of the lowest class:

```python
pd.crosstab(train['Pclass'], train['Embarked']).transform(lambda col: col / col.sum() * 100).round()
```

```python
train.groupby(['Pclass', 'Embarked']).agg(np.mean)
```

Let's replace all of the missings from Queenstown with the average value of the lower class from Queenstown, 26:

```python
age = train['Age']

mask = ~((train['Embarked']=='Q') & (train['Age'].isnull()))
age = age.where(mask, 26)

mask = ~((train['Embarked']=='S') & (train['Age'].isnull()))
age = age.where(mask, other=train[train['Embarked']=='S']['Age'].mean())

mask = ~((train['Embarked']=='C') & (train['Age'].isnull()))
age = age.where(mask, other=train[train['Embarked']=='C']['Age'].mean())

train['Age'] = age
```

Notice that for passengers from Southampton and Cherbourg, we have simply replaced the NaNs with mean age within those groups.

The only missingness remaining is in the cabin numbers, which are inherent in the data and cannot be imputed.

## Feature Engineering

Let's look into some feature engineering. First, can we identify at least some of the individuals as crew? Answer: No, because [this dataset only includes passengers](https://www.kaggle.com/c/titanic/data).

### `deck`

Let's extract the deck level from `Cabin`:

```python
train['deck'] = train['Cabin'].str[0]
train['deck'] = train['deck'].replace(np.nan, 'U')
# Cabin T was on the boat deck (top)
train['deck'] = pd.Categorical(train['deck'],
                               categories=['T', 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'U'],
                               ordered=True)
```

### `no_family`

Now let's examine the individual's family ties. First, let's create a binary variable `no_family` which measures if the individual was on the trip by themself:

```python
train['no_family'] = ((train['SibSp']==0) & (train['Parch']==0)).astype(int)
```

Let's also measure the total number of close family members that a passenger has aboard with them:

### `family_members`

```python
train['family_members'] = train['SibSp'] + train['Parch']
```

### `title`

Now, let's get passengers' titles. [This](https://triangleinequality.wordpress.com/2013/09/08/basic-feature-engineering-with-the-titanic-data/) guy already did some work for us:

```python
title_list = ['Mrs.', 'Mr.', 'Master.', 'Miss.', 'Major.', 'Rev.',
          'Dr.', 'Ms.', 'Mlle.','Col.', 'Capt.', 'Mme.', 'Countess.',
          'Don.', 'Jonkheer.']
```

```python
names = train['Name']
titles = list()
for name in names:
    for title in title_list:
        if title in name:
            titles.append(title)
            break
    else: titles.append(np.nan)
titles = pd.Series(titles)
train['title'] = titles
```

There are two passengers whose title doesn't fit in the list above:

```python
train[train['title'].isnull()]
```

```python
train.set_value(index=556, col='title', value='Mrs.')
train.set_value(599, 'title', 'Mr.')
```

Since on further inspection this is a couple, I decided to manually set their titles to "Mr." and "Mrs."

```python
train.title.unique()
```

```python
train.title.value_counts()
# Let's reduce the number of categories here
```

```python
# From the same source above
def replace_titles(x):
    title=x['title']
    if title in ['Don.', 'Major.', 'Capt.', 'Jonkheer.', 'Rev.', 'Col.']:
        return 'Mr.'
    elif title in ['Countess.', 'Mme.']:
        return 'Mrs.'
    elif title in ['Mlle.', 'Ms.']:
        return 'Miss.'
    elif title =='Dr.':
        if x['Sex']=='Male':
            return 'Mr.'
        else:
            return 'Mrs.'
    else:
        return title
    
train['title'] = train.apply(replace_titles, axis=1)
```

### `single`

Now, under the assumption that only unmarried boys and girls (or men and women) were given the titles 'Master' and 'Miss', let's infer passengers' marital status:

```python
single = train['title'].isin(['Master.', 'Miss.'])
train['single'] = single.astype(int)
```

### `maid`

There were several upper-class women with maids when I was perusing the passengers before. Let's formalize this and suppose that any woman going by the title "Miss" (implying that she is unmarried), with no family aboard, who is of adult age and in the upper class cabins, is a maid:

```python
maid = ((train['title'] == 'Miss.') 
        & (train['no_family']==1) 
        & (train['Age'] > 18) 
        & (train['Pclass']=='Upper'))
train['maid'] = maid.astype(int)
```

### `num_cabins`

In the course of this process, I noticed that some passengers reserved multiple cabins. Of course, this implies that this individual was a party leader and likely actually paid the fare for the trip.

```python
cabins = train.Cabin.str.split()
num_cabins = cabins.map(lambda x: len(x) if type(x)==list else np.nan)
train['num_cabins'] = num_cabins
train.head()
```

```python
train[train['num_cabins']>1].describe() - train.describe()
```

### `lname`, `lname_size`, `shares_lname`

Let's look at family groups by extracting and grouping last names:

```python
train['Name'].str.split?
```

```python
train['Name'].str.split?
```

```python
split_names = train['Name'].str.split(',')
lnames = split_names.map(lambda x: x[0])
train['lname'] = lnames

print(len(lnames.unique()))
```

We have 667 unique last names for 891 observations, for an average name group size of just over 1. Below, we see that the vast majority of last names are unique:

```python
train['lname'].value_counts().astype(str).value_counts().sort_index()
```

Let's add the features `lname_size` and `shares_lname`:

```python
lname_counts = train['lname'].value_counts()

lname_size = lname_counts[train['lname']]
lname_size.index = train['lname'].index
train['lname_size'] = lname_size
train['shares_lname'] = (train['lname_size'] > 1).astype(int)
```

### ~~`married_woman_name`~~

Married women have parentheses in their names. Let's add this as a feature and simultaneously check another feature, `married`:

```python
married_woman = train['Name'].map(lambda x: r'(' in x)
train['married_woman_name'] = married_woman.astype(int)
```

```python
women = train[train['Sex'] == 'female']
comparison_of_marriage_features = women[women['single'].astype(bool) != ~women['married_woman_name'].astype(bool)]
```

A visual inspection of the above indicates that the `single` variable is a better measure of a woman's marital status because the `married_woman_name` variable counts women with aliases or nicknames as married women, while the title-based `single` does not. We will drop `married_woman_name` and keep `single`:

```python
train = train.drop('married_woman_name', axis=1)
print(list(train.columns))
```

While we're at it, let's clean up our boolean variables and a few others:

```python
train[['Survived', 'no_family', 'maid', 'shares_lname', 'single']] = train[['Survived', 'no_family', 'maid', 'shares_lname', 'single']].astype(bool)
train[['married']] = ~train[['single']]
train = train.drop('single', axis=1)
```

```python
# Noticed some missings due to binning from before
train[['fare_binned']] = (train['fare_binned']
                          .where(cond=train['Fare']!=0, other='0-50'))
```
