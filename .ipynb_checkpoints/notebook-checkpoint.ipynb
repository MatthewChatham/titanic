## Data Dictionary

Looking at the data frame above, we see 12 columns. These are as follows:

* `PassengerID`: ID used for Kaggle competition
* `Survived`: 0 for deceased, 1 for survivor
* `Pclass`: The passenger class, where 1=First, 2=Second, and 3=Third (proxy for SES)
* `Name`: The passenger's name, of the form (last), (title) (first) (middle) ((wife's name))
* `Sex`: The passenger's sex, male and female
* `Age`: Passenger age in years
* `SibSp`: The number, combined, of siblings and spouses on board the ship
* `Parch`: The number, combined, of (blood) parents and children on board the ship
* `Ticket`: Ticket number
* `Fare`: Passenger fare
* `Cabin`: Passenger's cabin number (First Class only)
* `Embarked`: Point of embarkation, where S=Southampton, C=Cherboug, and Q=Queenstown

Let's transform the categorical variables:

## Clean up Column Data Types

```python
# Survived
df['Survived'] = df['Survived'].astype(bool)

# Pclass
df['Pclass'] = df['Pclass'].replace({1: 'Upper', 2: 'Middle', 3: 'Lower'})
df['Pclass'] = pd.Categorical(df['Pclass'], 
                                 categories=['Lower', 'Middle', 'Upper'], 
                                 ordered=True)
# Sex
df['Sex'] = pd.Categorical(df['Sex'], 
                                 categories=['male', 'female'], 
                                 ordered=False)
# Cabin
df['Cabin'] = pd.Categorical(df['Cabin'])

# Embarked
df['Embarked'] = pd.Categorical(df['Embarked'], 
                                 categories=['S', 'C', 'Q'], 
                                 ordered=True) # Order based on route
```

Let's look at the proportion of missing respondents by point of embarkation:

```python
def prop_missing(series):
    return series.isnull().sum() / series.isnull().size

df.groupby('Embarked').aggregate(prop_missing)[['Age', 'Cabin']].round(2)
```

Here we see that 60% of those who embarked at Queenstown have no age recorded, much higher than for those who embarked elsewhere. Below, we can see that 92% of those who embarked at Queenstown are from the lowest class.

```python
pd.crosstab(df['Pclass'], df['Embarked']).transform(lambda col: col / col.sum() * 100).round()
```

Given the above, let's replace all of the missings from Queenstown with the average value of the lower class from Queenstown, 26. We'll replace the missings from other points of embarkation with the average age for that group:

```python
age = df['Age']

mask = ~((df['Embarked']=='Q') & (df['Age'].isnull()))
age = df['Age'].where(mask, 26)

mask = ~((df['Embarked']=='S') & (df['Age'].isnull()))
age = age.where(mask, other=df[df['Embarked']=='S']['Age'].mean())

mask = ~((df['Embarked']=='C') & (df['Age'].isnull()))
age = age.where(mask, other=df[df['Embarked']=='C']['Age'].mean())

df['Age'] = age
```

## Binning `Age` and `Fare`

Let's bin age and fare for our decision tree farther down the road:

```python
# Bin `Age`
cut_points = [0, 10, 20, 30, 40, 50, 60, 70, 80]
lbls = ['{} - {}'.format(cut_points[i], cut_points[i+1]) 
        for i, val in enumerate(cut_points) if i < len(cut_points) - 1]
binned_age = pd.cut(df['Age'],
       bins=cut_points,
       labels=lbls)
df['age_binned'] = pd.Categorical(binned_age,
                                     categories=lbls, ordered=True)
# Bin `Fare`
cut_points = [i*50 for i in range(12)]
lbls = ['\${} - \${}'.format(cut_points[i], cut_points[i+1])
        for i, val in enumerate(cut_points) if i < len(cut_points) - 1]
binned_fare = pd.cut(df['Fare'],
       bins=cut_points,
       labels=lbls)
df['fare_binned'] = pd.Categorical(binned_fare,
                                     categories=lbls, ordered=True)
df['fare_binned'][ df['Fare']==0 ] = '\$0 - \$50'

df.head()
```

## Feature Engineering

Let's look into some feature engineering. First, can we identify at least some of the individuals as crew? Answer: No, because [this dataset only includes passengers](https://www.kaggle.com/c/titanic/data).

### `deck`

Let's extract the deck level from `Cabin`:

```python
df['deck'] = df['Cabin'].str[0]
df['deck'] = df['deck'].replace(np.nan, 'U')
# Cabin T was on the boat deck (top)
# "U" is unknown, which may be useful info
df['deck'] = pd.Categorical(df['deck'],
                               categories=['T', 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'U'],
                               ordered=True)
df.head()
```

### `no_family`

Now let's examine the individual's family ties. First, let's create a binary variable `no_family` which measures if the individual was on the trip by themself:

```python
df['no_family'] = ((df['SibSp']==0) & (df['Parch']==0)).astype(bool)
df.head()
```

Let's also measure the total number of close family members that a passenger has aboard with them:

### `total_family`

```python
df['total_family'] = df['SibSp'] + df['Parch']
df.head()
```

### `title`

Now, let's get passengers' titles. [This](https://triangleinequality.wordpress.com/2013/09/08/basic-feature-engineering-with-the-titanic-data/) guy already did some work for us:

```python
title_list = ['Mrs.', 'Mr.', 'Master.', 'Miss.', 'Major.', 'Rev.',
          'Dr.', 'Ms.', 'Mlle.','Col.', 'Capt.', 'Mme.', 'Countess.',
          'Don.', 'Jonkheer.']
```

```python
names = df['Name']
titles = list()
for name in names:
    for title in title_list:
        if title in name:
            titles.append(title)
            break
    else: titles.append(np.nan)
titles = pd.Series(titles)
titles.index = df.index
df['title'] = titles.astype('category')
```

There are three passengers whose title doesn't fit in the list above:

```python
df[df['title'].isnull()]
```

```python
df.set_value(index=('train', 556), col='title', value='Mrs.')
df.set_value(('train', 599), 'title', 'Mr.')
df.set_value(('test', 414), 'title', 'Mr.')
df[df['title'].isnull()]
```

The Dugg Gordons were a couple, so their titles are set to "Mr." and "Mrs.", while the third individual was of Spanish origin. "DoÃ±a" is Spanish for "Mrs."

```python
df.title.unique()
```

```python
df.title.value_counts()
# Let's reduce the number of categories here
```

```python
# From the same source above
def replace_titles(x):
    title=x['title']
    if title in ['Don.', 'Major.', 'Capt.', 'Jonkheer.', 'Rev.', 'Col.']:
        return 'Mr.'
    elif title in ['Countess.', 'Mme.']:
        return 'Mrs.'
    elif title in ['Mlle.', 'Ms.']:
        return 'Miss.'
    elif title =='Dr.':
        if x['Sex']=='Male':
            return 'Mr.'
        else:
            return 'Mrs.'
    else:
        return title
    
df.loc[:, 'title'] = df.apply(replace_titles, axis=1)
```

### `married`

Now, under the assumption that only unmarried boys and girls (or men and women) were given the titles 'Master' and 'Miss', let's infer passengers' marital status:

```python
married = df['title'].isin(['Mr.', 'Mrs.'])
df['married'] = married.astype(bool)
df.head()
```

### `is_maid`

There were several upper-class women with maids when I was perusing the passengers before. Let's formalize this and suppose that any woman going by the title "Miss" (implying that she is unmarried), with no family aboard, who is of adult age and in the upper class cabins, is a maid:

```python
maid = ((df['title'] == 'Miss.') 
        & (df['no_family']==1) 
        & (df['Age'] > 18) 
        & (df['Pclass']=='Upper'))
df['is_maid'] = maid.astype(bool)
df.head()
```

### `num_cabins`

In the course of this process, I noticed that some passengers reserved multiple cabins. This may indicate that this individual was a party leader and likely actually paid the fare for the trip.

```python
cabins = df.Cabin.str.split()
num_cabins = cabins.map(lambda x: len(x) if type(x)==list else np.nan)
df['num_cabins'] = num_cabins
df[df.num_cabins > 1].loc[:, ['Cabin', 'num_cabins']].head()
```

### `lname`, `lname_size`, `shares_lname`

Let's look at family groups by extracting and grouping last names:

```python
split_names = df['Name'].str.split(',')
lnames = split_names.map(lambda x: x[0])
df['lname'] = lnames
print(len(lnames.unique()))
df.head()
```

Below, we see that the vast majority of last names are unique:

```python
df['lname'].value_counts().astype(str).value_counts().sort_index()
```

Let's add the features `lname_size` and `shares_lname`:

```python
lname_counts = df['lname'].value_counts()
lname_size = lname_counts[df['lname']]
lname_size.index = df['lname'].index
df['lname_size'] = lname_size
df['shares_lname'] = (df['lname_size'] > 1).astype(bool)
df.head()
```

Before we run the model, we need to decide on what features to use and then prepare them for the decision tree classifier.

```python
feature_names = [
    'Pclass', 'Sex', 'SibSp', 'Parch',
    'age_binned', 'fare_binned', 'Embarked',
    'deck', 'no_family', 'total_family', 
    'title', 'married', 'is_maid', 'num_cabins', 
    'lname', 'lname_size', 'shares_lname'
]
raw_features = df[feature_names]
raw_features.head()
```

Let's convert some strings to categories.

```python
df['title'] = df['title'].astype('category')
df['lname'] = df['lname'].astype('category')
```

Let's one-hot encode our categorical variables:

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

le = LabelEncoder()
ohe = OneHotEncoder()

features = np.ndarray((1309,0))

pclass = ohe.fit_transform( le.fit_transform( raw_features['Pclass'] ).reshape(-1, 1) ).toarray()
sex = ohe.fit_transform( le.fit_transform( raw_features['Sex'] ).reshape(-1, 1) ).toarray()
embarked = ohe.fit_transform( le.fit_transform( raw_features['Embarked'] ).reshape(-1, 1) ).toarray()
age_binned = ohe.fit_transform( le.fit_transform( raw_features['age_binned'] ).reshape(-1, 1) ).toarray()
fare_binned = ohe.fit_transform( le.fit_transform( raw_features['fare_binned'] ).reshape(-1, 1) ).toarray()
deck = ohe.fit_transform( le.fit_transform( raw_features['deck'] ).reshape(-1, 1) ).toarray()
title = ohe.fit_transform( le.fit_transform( raw_features['title'] ).reshape(-1, 1) ).toarray()
lname = ohe.fit_transform( le.fit_transform( raw_features['lname'] ).reshape(-1, 1) ).toarray()

features = np.append(features, pclass, axis=1)
features = np.append(features, sex, axis=1)
features = np.append(features, embarked, axis=1)
features = np.append(features, age_binned, axis=1)
features = np.append(features, fare_binned, axis=1)
features = np.append(features, deck, axis=1)
features = np.append(features, title, axis=1)
features = np.append(features, lname, axis=1)
```
